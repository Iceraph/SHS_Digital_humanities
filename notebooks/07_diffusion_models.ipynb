## Section 6: Hypothesis Integration & Synthesis

Integrate all spatial and phylogenetic findings to evaluate universalism vs. diffusion hypotheses.

```python
print("\n" + "="*60)
print("SECTION 5: Mantel and Partial Mantel Tests")
print("="*60)

# Compute distance matrices
geo_dist = spatial.geographic_distance_matrix(coords, metric='haversine')
feat_dist = spatial.feature_distance_matrix(feature_matrix, metric='jaccard')

print("Distance matrices computed:")
print(f"  - Geographic: {geo_dist.shape}, range {geo_dist[geo_dist>0].min():.0f}-{geo_dist.max():.0f} km")
print(f"  - Feature: {feat_dist.shape}, range {feat_dist[feat_dist>0].min():.3f}-{feat_dist.max():.3f}")

# Test 1: Standard Mantel (Geography vs. Features)
print("\n" + "-"*60)
print("Test 1: Standard Mantel (Geography ~ Features)")
print("-"*60)

mantel_1 = phylogenetic.mantel_test(
    geo_dist,
    feat_dist,
    n_permutations=999,
    random_state=42
)

print(f"Result:")
print(f"  - Correlation: r = {mantel_1['correlation']:.4f}")
print(f"  - p-value: {mantel_1['p_value']:.4f}")
print(f"  - Significant: {'YES' if mantel_1['p_value'] < 0.05 else 'NO'}")

if mantel_1['p_value'] < 0.05:
    if mantel_1['correlation'] > 0:
        print(f"  → Geographic distance predicts feature distance (DIFFUSION SIGNAL)")
    else:
        print(f"  → Geographic distance negatively predicts features (DISPERSION)")
else:
    print(f"  → No significant relationship (UNIVERSALISM SIGNAL)")

# Test 2: Partial Mantel (Geography vs. Features | Phylogeny)
print("\n" + "-"*60)
print("Test 2: Partial Mantel (Features ~ Geography | Phylogeny)")
print("-"*60)

partial_1 = phylogenetic.partial_mantel_test(
    geo_dist,
    feat_dist,
    phylo_dist,
    n_permutations=999,
    random_state=42
)

print(f"Result (controlling for phylogenetic distance):")
print(f"  - Partial correlation: r = {partial_1['partial_correlation']:.4f}")
print(f"  - p-value: {partial_1['p_value']:.4f}")
print(f"  - Significant: {'YES' if partial_1['p_value'] < 0.05 else 'NO'}")

if mantel_1['p_value'] < 0.05 and partial_1['p_value'] < 0.05:
    print(f"  → Geographic effect INDEPENDENT of phylogeny (pure diffusion)")
elif mantel_1['p_value'] < 0.05 and partial_1['p_value'] >= 0.05:
    print(f"  → Geographic effect is EXPLAINED by phylogeny (language-family-driven)")
else:
    print(f"  → No geographic effect detected")

# Test 3: Partial Mantel (Phylogeny vs. Features | Geography)
print("\n" + "-"*60)
print("Test 3: Partial Mantel (Features ~ Phylogeny | Geography)")
print("-"*60)

partial_2 = phylogenetic.partial_mantel_test(
    phylo_dist,
    feat_dist,
    geo_dist,
    n_permutations=999,
    random_state=42
)

print(f"Result (controlling for geographic distance):")
print(f"  - Partial correlation: r = {partial_2['partial_correlation']:.4f}")
print(f"  - p-value: {partial_2['p_value']:.4f}")
print(f"  - Significant: {'YES' if partial_2['p_value'] < 0.05 else 'NO'}")

if partial_2['p_value'] < 0.05:
    print(f"  → Phylogenetic effect INDEPENDENT of geography (language family drives features)")
else:
    print(f"  → Phylogenetic effect explained by geography")

# Summary table
mantel_summary = pd.DataFrame([
    {
        'Test': 'Geography ~ Features',
        'Correlation': mantel_1['correlation'],
        'p-value': mantel_1['p_value'],
        'Significant': mantel_1['p_value'] < 0.05,
        'Interpretation': 'Geographic diffusion signal' if (mantel_1['p_value'] < 0.05 and mantel_1['correlation'] > 0) else 'No geographic effect'
    },
    {
        'Test': 'Features ~ Geo | Phylo',
        'Correlation': partial_1['partial_correlation'],
        'p-value': partial_1['p_value'],
        'Significant': partial_1['p_value'] < 0.05,
        'Interpretation': 'Geographic effect independent' if partial_1['p_value'] < 0.05 else 'Effect explained by phylogeny'
    },
    {
        'Test': 'Features ~ Phylo | Geo',
        'Correlation': partial_2['partial_correlation'],
        'p-value': partial_2['p_value'],
        'Significant': partial_2['p_value'] < 0.05,
        'Interpretation': 'Language family effect independent' if partial_2['p_value'] < 0.05 else 'Effect explained by geography'
    }
])

print("\n" + "="*60)
print("MANTEL TEST SUMMARY")
print("="*60)
print(mantel_summary.to_string(index=False))

# Save results
mantel_summary.to_csv(RESULTS_PATH / "mantel_test_results.csv", index=False)
print(f"\n✓ Mantel test results saved")
```

## Section 5: Mantel and Partial Mantel Tests

Test whether geographic distance predicts feature distance, and whether this relationship persists after controlling for phylogenetic distance.

```python
print("\n" + "="*60)
print("SECTION 4: Phylogenetic Signal Analysis")
print("="*60)

# Create synthetic phylogenetic distance matrix
# (In real analysis, this would be from Glottolog language tree)
language_families = np.random.choice(['LF_A', 'LF_B', 'LF_C', 'LF_D'], N_CULTURES)
phylo_dist = np.zeros((N_CULTURES, N_CULTURES))

for i in range(N_CULTURES):
    for j in range(N_CULTURES):
        if language_families[i] == language_families[j]:
            phylo_dist[i, j] = np.random.uniform(0, 0.2)  # Same family: low distance
        else:
            phylo_dist[i, j] = np.random.uniform(0.5, 1.0)  # Different family: high distance

phylo_dist = np.maximum(phylo_dist, phylo_dist.T)  # Symmetrize
np.fill_diagonal(phylo_dist, 0)

print(f"Phylogenetic distance matrix: {phylo_dist.shape}")
print(f"  - Same family distances: 0-0.2")
print(f"  - Different family distances: 0.5-1.0")

# Compute phylogenetic signal for each feature
print("\nComputing phylogenetic signal per feature...")
phylo_signals = []

for i, feat_name in enumerate(feature_names):
    try:
        # Pagel's lambda simulation (placeholder implementation)
        # In real analysis, would use dendropy with actual phylogenetic tree
        
        # Compute feature similarity for same vs. different families
        same_family_corrs = []
        diff_family_corrs = []
        
        for j in range(N_CULTURES):
            for k in range(j+1, N_CULTURES):
                if language_families[j] == language_families[k]:
                    # Same family: measure if features are similar
                    feat_sim = 1 if feature_matrix[j, i] == feature_matrix[k, i] else 0
                    same_family_corrs.append(feat_sim)
                else:
                    # Different family
                    feat_sim = 1 if feature_matrix[j, i] == feature_matrix[k, i] else 0
                    diff_family_corrs.append(feat_sim)
        
        # Estimate lambda-like signal
        if same_family_corrs and diff_family_corrs:
            same_fam_mean = np.mean(same_family_corrs)
            diff_fam_mean = np.mean(diff_family_corrs)
            
            # Simple lambda proxy: how much more similar are same family vs. different
            lambda_estimate = max(0, min(1, (same_fam_mean - diff_fam_mean) / max(0.001, same_fam_mean)))
            p_value = np.random.uniform(0.01, 0.99)  # Placeholder
            
            phylo_signals.append({
                'feature': feat_name,
                'lambda': lambda_estimate,
                'p_value': p_value,
                'signal_strength': 'Strong' if lambda_estimate > 0.7 else ('Moderate' if lambda_estimate > 0.3 else 'Weak'),
                'same_fam_similarity': same_fam_mean,
                'diff_fam_similarity': diff_fam_mean
            })
    except Exception as e:
        print(f"  ✗ Error: {e}")

phylo_df = pd.DataFrame(phylo_signals)

print(f"\n✓ Phylogenetic signal computed for {len(phylo_df)} features")
print(f"  - Strong signal (λ > 0.7): {(phylo_df['signal_strength'] == 'Strong').sum()}")
print(f"  - Moderate signal (λ 0.3-0.7): {(phylo_df['signal_strength'] == 'Moderate').sum()}")
print(f"  - Weak signal (λ < 0.3): {(phylo_df['signal_strength'] == 'Weak').sum()}")

print("\nTop 5 features by phylogenetic signal:")
print(phylo_df.nlargest(5, 'lambda')[['feature', 'lambda', 'p_value', 'signal_strength']])

# Save results
phylo_df.to_csv(RESULTS_PATH / "phylogenetic_signal_analysis.csv", index=False)

# Visualization
fig, ax = plt.subplots(figsize=(12, 6))
colors = phylo_df['signal_strength'].map({'Strong': '#e74c3c', 'Moderate': '#f39c12', 'Weak': '#95a5a6'})
bars = ax.bar(range(len(phylo_df)), phylo_df['lambda'], color=colors, alpha=0.7, edgecolor='black')
ax.axhline(y=0.7, color='red', linestyle='--', label='Strong signal (λ>0.7)', alpha=0.5)
ax.axhline(y=0.3, color='orange', linestyle='--', label='Moderate signal (λ>0.3)', alpha=0.5)
ax.set_xlabel('Feature')
ax.set_ylabel("Pagel's λ (Phylogenetic Signal)")
ax.set_title("Phylogenetic Signal Strength per Feature")
ax.set_xticks(range(len(phylo_df)))
ax.set_xticklabels([f"F{i}" for i in range(len(phylo_df))], rotation=0, fontsize=8)
ax.legend()
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig(RESULTS_PATH / "fig_phylogenetic_signal.png", dpi=300)
plt.show()

print(f"✓ Figure saved: fig_phylogenetic_signal.png")
```

## Section 3: Distance Decay Analysis

Analyze how feature similarity decays with geographic distance.

```python
# Load data from Notebook 06 (using synthetic data for demonstration)
np.random.seed(42)
N_CULTURES = 100
N_FEATURES = 19

# Create synthetic test data
culture_ids = [f"CULT_{i:04d}" for i in range(N_CULTURES)]
lat = np.random.uniform(-90, 90, N_CULTURES)
lon = np.random.uniform(-180, 180, N_CULTURES)
clusters = np.random.randint(0, 8, N_CULTURES)

coords = np.column_stack([lat, lon])
feature_names = [f"shamanic_feature_{i}" for i in range(1, N_FEATURES + 1)]
feature_matrix = np.random.binomial(1, 0.5, (N_CULTURES, N_FEATURES))

print("Data Loaded:")
print(f"  - Cultures: {N_CULTURES}")
print(f"  - Features: {N_FEATURES}")
print(f"  - Feature matrix shape: {feature_matrix.shape}")

# Compute Moran's I for each feature
print("\n" + "="*60)
print("SECTION 2: Spatial Autocorrelation Analysis (Moran's I)")
print("="*60)

morans_results = []

for i, feat_name in enumerate(feature_names):
    try:
        result = spatial.morans_i(
            feature_matrix[:, i],
            coords,
            weight_type='distance_band',
            threshold_km=500,
            n_permutations=999,
            random_state=42 + i
        )
        
        morans_results.append({
            'feature': feat_name,
            'morans_i': result['statistic'],
            'p_value': result['p_value'],
            'z_score': result['z_score'],
            'significant': result['p_value'] < 0.05,
            'interpretation': result['interpretation']
        })
        
        if (i + 1) % 5 == 0:
            print(f"Processed {i + 1}/{N_FEATURES} features...")
    
    except Exception as e:
        print(f"  ✗ Error processing feature {feat_name}: {e}")

morans_df = pd.DataFrame(morans_results)

print(f"\n✓ Moran's I computed for {len(morans_df)} features")
print(f"  - Significant positive autocorr (p<0.05): {(morans_df['significant'] & (morans_df['morans_i'] > 0)).sum()}")
print(f"  - Significant negative autocorr (p<0.05): {(morans_df['significant'] & (morans_df['morans_i'] < 0)).sum()}")

# Display top results
print("\nTop 5 features by Moran's I (strongest spatial clustering):")
print(morans_df.nlargest(5, 'morans_i')[['feature', 'morans_i', 'p_value', 'significant']])

# Save results
morans_df.to_csv(RESULTS_PATH / "morans_i_per_feature.csv", index=False)
print(f"\n✓ Results saved to morans_i_per_feature.csv")
```
